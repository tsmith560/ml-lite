```python
import pandas as pd
```

```python
swag_df = pd.read_csv("data/swag.csv")
```

```python
print(swag_df.head())
```

```python
import random
```

```python
def recommend_random_swag(df):
    item = df.sample(1).iloc[0]
    return {
        "name": item["name"],
        "description": item["description"],
        "image_path": item["image_path"]
    }
```

```python
swag = recommend_random_swag(swag_df)
print(swag)
```

```python
beer_df = pd.read_csv("data/beer.csv")
```

```python
print(beer_df.head())
```

```python
import random
```

```python
def recommend_random_beer(df):
    item = df.sample(1).iloc[0]
    return {
        "name": item["name"],
        "tagline": item["tagline"],
        "abv": item["abv"],
    	"origin": item["origin"]
    }
```

```python
beer = recommend_random_beer(beer_df)
print(beer)
```

```python
def recommend_random_song(df):
    item = df.sample(1).iloc[0]
    return {
        "title": item["title"],
        "artist": item["artist"],
        "spotify_url": item["spotify_url"]
    }
```

```python
import ipywidgets as widgets
from IPython.display import display

# Use Jupyter widgets to create interactive window for testing
all_tags = sorted(set(
    tag.strip().lower()
    for tags_list in beer_df['tags'].dropna()
    for tag in tags_list.split(',')
))

tag_selector = widgets.SelectMultiple(
    options=all_tags,
    description='Tags',
    disabled=False
)

button = widgets.Button(description="Generate Miller Moment")

output = widgets.Output()

def on_button_click(b):
    with output:
        output.clear_output()
        selected_tags = list(tag_selector.value)
        moment = generate_miller_moment(beer_df, swag_df, song_df, tags=selected_tags)
        print(f"üç∫ Beer: {moment['beer']['name']} ‚Äî {moment['beer']['tagline']} ({moment['beer']['abv']}% ABV, {moment['beer']['origin']})")
        print(f"üéÅ Swag: {moment['swag']['name']} ‚Äî {moment['swag']['description']}")
        print(f"üéµ Song: {moment['song']['title']} by {moment['song']['artist']} ‚Äî {moment['song']['spotify_url']}")
        display(Image(filename=moment['swag']['image_path']))

button.on_click(on_button_click)

display(tag_selector, button, output)
```

```python
# Simple input method for testing
user_input = input("Enter tags separated by commas (e.g. summer, bbq, classic): ")
tags = [tag.strip().lower() for tag in user_input.split(",") if tag.strip()]

moment = generate_miller_moment(beer_df, swag_df, song_df, tags=tags)

print(f"üç∫ Beer: {moment['beer']['name']} ‚Äî {moment['beer']['tagline']} ({moment['beer']['abv']}% ABV, {moment['beer']['origin']})")
print(f"üéÅ Swag: {moment['swag']['name']} ‚Äî {moment['swag']['description']}")
print(f"üéµ Song: {moment['song']['title']} by {moment['song']['artist']} ‚Äî {moment['song']['spotify_url']}")

from IPython.display import Image, display
display(Image(filename=moment['swag']['image_path']))
```
